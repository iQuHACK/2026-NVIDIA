```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
benchmark_data = pd.read_csv('benchmark_results.csv')
benchmark_data.head()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(benchmark_data['N'], benchmark_data['Conv. Time'], label='Conv. Time')
plt.scatter(benchmark_data['N'], benchmark_data['Total Time'], label='Total Time')
plt.xlabel('N')
plt.ylabel('Time')
plt.legend()
plt.show()
```
